Q2. What is the main advantage of using Lasso Regression in feature selection?

The main advantage is that Lasso can shrink some coefficients exactly to zero.

This means:

✔ it automatically selects important features
✔ it produces a simpler and more interpretable model

Q4. What are the tuning parameters in Lasso Regression and how do they affect performance?

The main tuning parameter is:

✅ λ (lambda or alpha)

controls the strength of regularization

λ value	Effect
Small λ	Model behaves close to ordinary linear regression
Large λ	More coefficients become zero → simpler model

Trade-off:

small λ → low bias, high variance

large λ → higher bias, lower variance

Q6. What is the difference between Ridge Regression and Lasso Regression?

| Aspect | Ridge | Lasso |
-----|------
Penalty | L2 (squared coefficients) | L1 (absolute coefficients)
Coefficient shrinkage | Shrinks toward zero | Shrinks and can set to zero
Feature selection | No | Yes
When predictors are correlated | Keeps all predictors | Tends to select one and drop others

Q7. Can Lasso Regression handle multicollinearity? If yes, how?

Yes, Lasso can handle multicollinearity, but in a different way from Ridge.

When predictors are highly correlated:

Lasso often keeps only one variable from the group

and sets the others to zero

This is useful for feature selection, but:

the choice of which correlated variable is selected can be unstable.

So Lasso handles multicollinearity by selecting among correlated predictors, not by stabilizing them.

Q8. How do you choose the optimal value of λ in Lasso Regression?

The standard approach is:

✅ Cross-validation

Steps:

Define a grid of λ values

Perform k-fold cross-validation

Select the λ with the lowest validation error (RMSE, MAE, or MSE)

In practice, tools such as LassoCV (scikit-learn) are commonly used.

✅ Final summary

Lasso is mainly used when:

many predictors exist

feature selection is important

a sparse model is preferred
